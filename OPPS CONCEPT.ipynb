```python
# Creating base class - Account : 

class Account:
    
 def __init__(self,balance):
        if balance >= 0.0:
            self.balance = balance
        else:
            self.balance = 0.0
            print("intial balance was invalid")
            
            
 def debit(self,damt):
        if damt > self.balance:
            print("Debit amount exceeded account balance")
            return False
        else:
            self.balance = self.balance - damt       # damt = debit amount
            return True
            
            
 def credit(self,camt):
        self.balance = self.balance + camt          # camt = credit amount 
        
               
 def getbalance(self):
        return self.balance
            
  
```

```python
# Derived Class: SavingsAccount
class SavingsAccount(Account):
    
    def __init__(self, balance, roi):
        super().__init__(balance)
        self.roi = roi                         # roi = Rate of interest

    def calculate_interest(self):
        return (self.balance * 1 * self.roi) / 100
        


# Derived Class: CheckingAccount

class CheckingAccount(Account):
    def __init__(self, balance, tfee):
        super().__init__(balance)
        self.tfee = tfee                       # tfee = transaction fee

    def debit(self, damt):
        res = super().debit(damt)
        if res:
            self.balance = self.balance - self.tfee
            return res
            
    def credit(self, camt):
        super().credit(camt)
        self.balance = self.balance - self.tfee

    
            


# Derived Class: CurrentAccount

class CurrentAccount(Account):
    
    def __init__(self, balance, over_draft):
        super().__init__(balance)
        self.over_draft = over_draft

    def debit(self, damt):
        if damt > (self.balance + self.over_draft):
            print("Debit amount exceeded account balance")
            return False
        else:
            self.balance = self.balance - damt
            return True
```
